```python
data = pd.read_csv('data/power.csv')
```

```python
# find the names of the following countries Латвия, Литва и Эстония

data[data['country'].str.contains(r'^(La)|^(Li)|^(Est)', regex=True)]['country'].unique()
```

```python
countries = data['country']
year = data['year']
category = data['category']

country_filter = (countries =='Lithuania') | (countries =='Latvia') |(countries =='Estonia')

year_filter = ((2005 <= year) & (year >= 2010))

category_filter = (category == 4) | (category == 12) | (category == 21)
```

```python
filtered_countries = data[country_filter & year_filter & category_filter & (data['quantity'] >= 0)]
filtered_countries['country'].unique()
```

```python
filtered_countries
```

```python
sum(filtered_countries['quantity'].values)
```
